````python
# CELL X — Setup Gemini LLM API

!pip install -q google-generativeai

import os
import google.generativeai as genai
from google.colab import userdata

print("google-generativeai installed. Configuring Gemini...")

# Retrieve the Google API Key securely
GOOGLE_API_KEY = None
try:
    GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
    print("Google API Key loaded from Colab Secrets.")
except userdata.notebook_utils.NotebookException:
    print("Warning: GOOGLE_API_KEY not found in Colab Secrets. Checking environment variables.")

if not GOOGLE_API_KEY:
    # Fallback for local testing or non-Colab environments
    # !pip install python-dotenv
    # from dotenv import load_dotenv
    # load_dotenv()
    GOOGLE_API_KEY = os.getenv('GOOGLE_API_KEY')
    if GOOGLE_API_KEY:
        print("Google API Key loaded from environment variables.")

if not GOOGLE_API_KEY:
    raise ValueError("GOOGLE_API_KEY not found. Please set it in Colab Secrets or as an environment variable.")

# Configure the generative AI model
genai.configure(api_key=GOOGLE_API_KEY);

# List available models to debug model availability
available_models = [m for m in genai.list_models() if "generateContent" in m.supported_generation_methods]
print("Available Gemini Models for generateContent:")
for m in available_models:
    print(f"  - {m.name}")

# Initialize the Gemini model with a more specific model name
# Prioritize gemini-1.5-flash, fall back to gemini-flash-latest, then gemini-pro if not available
model_name = 'gemini-1.5-flash'
if f"models/{model_name}" not in [m.name for m in available_models]:
    print(f"Warning: {model_name} not found. Attempting to use 'gemini-flash-latest'.")
    model_name = 'gemini-flash-latest'
    if f"models/{model_name}" not in [m.name for m in available_models]:
        print(f"Warning: {model_name} not found. Falling back to 'gemini-pro'.")
        model_name = 'gemini-pro'

model = genai.GenerativeModel(model_name)

print(f"Gemini model '{model_name}' initialized successfully!")
````

````python
import fitz, os, json, re
from tqdm import tqdm
from datetime import datetime
import concurrent.futures # Import for parallel processing

# Assuming `model` (the Gemini GenerativeModel) is initialized from a previous cell.
# If this cell is run independently, ensure the Gemini setup cell is run first.

output_file = "/content/drive/MyDrive/SRI_LANKA_CASE_DATABASE.json"

# Keep track of results in a dictionary for easy updating of retries
final_results_map = {}
# Keep track of successfully processed files for resume logic
successfully_processed_files_set = set()

# --- Resume functionality ---
# 1. Try loading the main output file
if os.path.exists(output_file):
    with open(output_file, "r", encoding="utf-8") as f:
        try:
            loaded_results = json.load(f)
            for item in loaded_results:
                if item.get("source_file"):
                    final_results_map[item["source_file"]] = item
                    if "error" not in item: # Only add to set if no error
                        successfully_processed_files_set.add(item["source_file"])
            print(f"Loaded {len(loaded_results)} existing results from {output_file}. Found {len(successfully_processed_files_set)} successful extractions.")
        except json.JSONDecodeError:
            print(f"Error decoding JSON from {output_file}. Starting fresh.")

# 2. Check for backup files and load the latest if no results loaded or if main output only had errors
# Only load backup if 'final_results_map' is empty OR if all existing results in map are errors.
if not final_results_map or all("error" in item for item in final_results_map.values()):
    backup_files_dir = "/content/drive/MyDrive"
    backup_files = sorted([f for f in os.listdir(backup_files_dir) if f.startswith("backup_") and f.endswith(".json")], reverse=True)

    if backup_files:
        latest_backup = os.path.join(backup_files_dir, backup_files[0])
        with open(latest_backup, "r", encoding="utf-8") as f:
            try:
                backup_loaded_results = json.load(f)
                # Clear existing results if we are replacing them with a backup
                if final_results_map and all("error" in item for item in final_results_map.values()):
                    final_results_map.clear()
                    successfully_processed_files_set.clear()

                for item in backup_loaded_results:
                    if item.get("source_file"):
                        final_results_map[item["source_file"]] = item
                        if "error" not in item:
                            successfully_processed_files_set.add(item["source_file"])
                print(f"Loaded {len(backup_loaded_results)} existing results from latest backup {latest_backup}. Found {len(successfully_processed_files_set)} successful extractions.")
            except json.JSONDecodeError:
                print(f"Error decoding JSON from backup {latest_backup}. Continuing with current state.")

# Get all PDF files and filter out already successfully processed ones
all_pdf_files = [f for f in os.listdir(PDF_FOLDER) if f.lower().endswith('.pdf')][:4000] # limit if needed

# Create a mapping from filename to its original index for case_id consistency
filename_to_original_idx = {filename: idx for idx, filename in enumerate(all_pdf_files)}

# Filter PDFs to process: only those not in successfully_processed_files_set
files_to_process = []
for filename in all_pdf_files:
    if filename not in successfully_processed_files_set:
        files_to_process.append((filename, filename_to_original_idx[filename]))

print(f"Found {len(all_pdf_files)} total PDFs. {len(successfully_processed_files_set)} successfully processed previously. Processing {len(files_to_process)} remaining PDFs.\n")

def process_pdf(filename, original_idx, folder):
    """Processes a single PDF file and returns extracted data or an error."""
    path = os.path.join(folder, filename)
    # Use original_idx to maintain consistent case_id sequencing
    case_id_prefix = f"SL-{original_idx+1001:04d}"

    print(f"--- Starting processing for {filename} (Case ID: {case_id_prefix}) ---")

    try:
        doc = fitz.open(path)
        text = "\n".join(page.get_text() for page in doc)

        prompt = f"""
You are a Supreme Court judge of Sri Lanka with 40 years experience.
Read the FULL judgment and classify it professionally.

Return ONLY valid JSON. No extra text.

{{
  "case_id": "SL-{original_idx+1001:04d}",
  "case_name": "Exact full case name from first page",
  "main_category": "Choose ONE from these only: Employment | Lease/Tenancy | Commercial Contract | Loan Agreement | Public Procurement | Tax Agreement | Debt Recovery | Trust & Property Transfer",
  "sub_category": "Add a short specific sub-type if needed (e.g., Termination, Short Notice, Gratuity, Unjust Enrichment, Ultra Vires, Prescription)",
  "is_contract_related": true or false,
  "judge_statement": ["Copy the EXACT 1-3 sentences where the judge says the contract clause/action was illegal/unfair. Use bullet points."],
  "violation_summary": "In 2-4 clear sentences: What exactly went wrong in the contract that caused loss?",
  "law_cited": ["List only the MAIN laws/acts violated"],
  "outcome": "Who won + what was awarded (e.g., Employee won: reinstated + 36 months salary + gratuity + costs)",
  "key_takeaway": "One sentence or in points: What must companies avoid in contracts after this case?"
}}

Examples of good main_category:
- Employment
- Lease/Tenancy
- Commercial Contract
- Public Procurement
- Trust & Property Transfer

Judgment (READ EVERY WORD):
{text[:42000]}
"""
        # Use the global 'model' variable which was initialized in the Gemini setup cell
        response = model.generate_content(prompt)
        raw = response.text.strip()

        # Debugging: Print raw response
        if not raw:
            raise ValueError("Empty response from model.")

        if "```json" in raw:
            raw = raw.split("```json")[1].split("```")[0]

        data = {}
        try:
            data = json.loads(raw)
        except json.JSONDecodeError as e:
            print(f"JSON Decode Error for {filename}: {e}\nRaw response: {raw[:500]}...") # Print partial raw response
            raise e # Re-raise to be caught by outer except

        data["source_file"] = filename
        data["extracted_at"] = datetime.now().isoformat()

        # Use .get() to safely access keys, providing a default if missing
        print(f"SUCCESS {data.get('case_id', 'N/A')} | {data.get('main_category', 'N/A')} | {data.get('violation_summary', 'N/A')[:60]}...")
        return data

    except Exception as e:
        error_message = f"FAILED {filename} (Case ID: {case_id_prefix}): {e}"
        print(error_message)
        return {"case_id": case_id_prefix, "case_name": "ERROR", "source_file": filename, "error": str(e)}

# Main execution using ThreadPoolExecutor
MAX_WORKERS = min(32, os.cpu_count() + 4)

with concurrent.futures.ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
    # Submit tasks only for the files that need processing (not successfully processed before)
    futures = {executor.submit(process_pdf, filename, original_idx, PDF_FOLDER): (filename, original_idx) for filename, original_idx in files_to_process}

    # Iterate as tasks complete, displaying progress with tqdm
    for future in tqdm(concurrent.futures.as_completed(futures), total=len(files_to_process), desc="Overall Processing"):
        result = future.result()
        # Update the map with the latest result for this file (overwriting previous errors if successful)
        final_results_map[result["source_file"]] = result

# Convert the map back to a list for saving
final_results_list = list(final_results_map.values())

# FINAL SAVE
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(final_results_list, f, indent=2, ensure_ascii=False)

print(f"\nDONE! {len(final_results_list)} cases saved to:\n{output_file}")
````

````python
# preview_before_db.py  ← RUN THIS NOW
import json
from collections import Counter
import os # Import os for path checking

output_file = "/content/drive/MyDrive/SRI_LANKA_CASE_DATABASE.json"

# Debugging: List contents of the directory
# print(f"Checking directory: {os.path.dirname(output_file)}")
# !ls -l {os.path.dirname(output_file)}
# print("-" * 30)

# Check if the file exists before attempting to open it
if not os.path.exists(output_file):
    raise FileNotFoundError(f"The file '{output_file}' was not found. Please ensure the previous PDF extraction cell (e.g., CELL 4) was run successfully and the file exists in your Google Drive.")

# Load extracted data
with open(output_file, "r", encoding="utf-8") as f:
    data = json.load(f)

print(f"TOTAL CASES EXTRACTED: {len(data)}\n")
print("="*80)

# 1. How many are actually contract-related?
contract_cases = [c for c in data if c.get("is_contract_related") == True]
non_contract = len(data) - len(contract_cases)

print(f"CONTRACT-RELATED CASES (will go to DB): {len(contract_cases)}")
print(f"NON-CONTRACT CASES (will be filtered out): {non_contract}\n")

# 2. Show all unique domains that will become contract_types table
main_categories = [c["main_category"] for c in contract_cases if c.get("main_category")]
main_category_counts = Counter(main_categories)
print("DOMAINS → contract_types table will have these:")
for main_category, count in main_category_counts.most_common():
    print(f"   • {main_category:35} → {count} cases")
print()

# 2. Show all unique domains that will become contract_types table
sub_categories = [c["sub_category"] for c in contract_cases if c.get("sub_category")]
sub_category_counts = Counter(sub_categories)
print("Sub categories → contract_types table will have these:")
for sub_category, count in sub_category_counts.most_common():
    print(f"   • {sub_category:35} → {count} cases")
print()

# 3. Sample of real judge statements
print("SAMPLE OF REAL JUDGE STATEMENTS (these become warnings in legal_rules):")
for i, case in enumerate(contract_cases[:8]):
    name = case["case_name"][:70] + "..." if len(case["case_name"]) > 70 else case["case_name"]
    statement = case.get("judge_statement", "")
    if isinstance(statement, list):
        statement = " | ".join(statement[:2])
    print(f"{case['case_id']} | {name}")
    print(f"   → {statement[:140]}{'...' if len(statement)>140 else ''}\n")

# 4. How many unique laws will become rules?
all_laws = []
for case in contract_cases:
    laws = case.get("law_cited", [])
    if isinstance(laws, str):
        laws = [laws]
    all_laws.extend([law.strip() for law in laws if law.strip()])

print(f"TOTAL UNIQUE LAWS/RULES that will go into legal_rules table: {len(set(all_laws))}")
print("Top 15 most common laws cited:")
for law, count in Counter(all_laws).most_common(15):
    print(f"   • {count:3} × {law}")

print("\n" + "="*80)
````
